```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
user = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.11'}
url = 'https://books.toscrape.com/'
webpage = requests.get(url,headers= user).text
```

```python
webpage
```

```python
soup = BeautifulSoup(webpage,"lxml")
```

```python
print(soup.prettify())
```

```python
soup.find_all('title')
```

```python
len(soup.find_all('h3')) #returns no. of books in 1 page
```

```python
soup.find_all('h3')[3].text
```

```python
books=[]
for j in range(1,51):
  for i in soup.find_all('article', class_='product_pod'):
    title = i.h3.a['title']
    books.append(title)
```

```python
len(books)
```

```python
books
```

```python
price = []
for j in range(1,51):
  for i in soup.find_all('p',class_='price_color'):
   price.append(i.text)
```

```python
len(price)
```

```python
availability =[]
for j in range(1,51):
  for i in soup.find_all('p', class_='instock availability'):
    availability.append(i.text.strip())
```

```python
len(availability)
```

```python
import pandas as pd
data = {'Title':books,'Price':price,'Availability':availability}
df = pd.DataFrame(data)
```
